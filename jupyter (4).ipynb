```python
# Password Strength Analyzer Project
print(" Password Strength Analyzer")
```

```python
import re
```

```python
password = input("Enter your password: ")
```

```python
score = 0
```

```python
if len(password) >= 8:
    score += 1
else:
    print(" Password should be at least 8 characters long")
```

```python
if re.search("[A-Z]", password):
    score += 1
else:
    print(" Add at least one uppercase letter")
```

```python
if re.search("[a-z]", password):
    score += 1
else:
    print(" Add at least one lowercase letter")
```

```python
if re.search("[0-9]", password):
    score += 1
else:
    print(" Add at least one number")
```

```python
if re.search("[@#$%^&*]", password):
    score += 1
else:
    print(" Add at least one special character")
```

```python
common_passwords = ["123456", "password", "12345678", "qwerty"]

if password in common_passwords:
    print(" This password is very common and unsafe")
```

```python
print("\nPassword Strength:")

if score <= 2:
    print(" Weak Password")
elif score == 3 or score == 4:
    print(" Medium Password")
else:
    print(" Strong Password")

print(f"Score: {score}/5")
```

```python
import pickle

# Create a "model" (rules/config)
password_model = {
    "min_length": 8,
    "require_upper": True,
    "require_lower": True,
    "require_digit": True,
    "require_special": True,
    "special_chars": "@#$%^&*",
    "common_passwords": ["123456", "password", "qwerty", "12345678"]
}

# Save model to file
with open("password_model.pkl", "wb") as file:
    pickle.dump(password_model, file)

print("✅ Model saved as password_model.pkl")
```
