```python
# ===============================
# PASSWORD STRENGTH ANALYZER
# ===============================

import re
import hashlib
import random
import string
import pickle
import os

# -------------------------------
# File to store password database
# -------------------------------
DB_FILE = "password_db.pkl"

# -------------------------------
# Load database from pickle
# -------------------------------
def load_database():
    if os.path.exists(DB_FILE):
        with open(DB_FILE, "rb") as f:
            return pickle.load(f)
    return set()

# -------------------------------
# Save database to pickle
# -------------------------------
def save_database(db):
    with open(DB_FILE, "wb") as f:
        pickle.dump(db, f)

# Initialize database
password_db = load_database()

# -------------------------------
# Function: Hash Password
# -------------------------------
def hash_password(password):
    return hashlib.sha256(password.encode()).hexdigest()

# -------------------------------
# Function: Check Password Strength
# -------------------------------
def check_strength(password):
    score = 0
    remarks = []

    # Length Check
    if len(password) >= 12:
        score += 2
    elif len(password) >= 8:
        score += 1
    else:
        remarks.append("Password too short")

    # Uppercase
    if re.search(r"[A-Z]", password):
        score += 1
    else:
        remarks.append("Add uppercase letters")

    # Lowercase
    if re.search(r"[a-z]", password):
        score += 1
    else:
        remarks.append("Add lowercase letters")

    # Digits
    if re.search(r"\d", password):
        score += 1
    else:
        remarks.append("Add numbers")

    # Special Characters
    if re.search(r"[!@#$%^&*(),.?\":{}|<>]", password):
        score += 1
    else:
        remarks.append("Add special characters")

    # Final Strength
    if score >= 6:
        strength = "Strong"
    elif score >= 4:
        strength = "Medium"
    else:
        strength = "Weak"

    return strength, remarks

# -------------------------------
# Function: Check Uniqueness
# -------------------------------
def is_unique(password):
    hashed = hash_password(password)
    return hashed not in password_db

# -------------------------------
# Function: Generate Strong Password
# -------------------------------
def generate_password(length=12):
    all_chars = string.ascii_letters + string.digits + "!@#$%^&*()"
    return ''.join(random.choice(all_chars) for _ in range(length))

# -------------------------------
# Function: Add Password to DB
# -------------------------------
def store_password(password):
    hashed = hash_password(password)
    password_db.add(hashed)
    save_database(password_db)

# -------------------------------
# MAIN PROGRAM
# -------------------------------
def main():
    print("\n🔐 PASSWORD STRENGTH ANALYZER\n")

    password = input("Enter your password: ")

    # Strength Check
    strength, remarks = check_strength(password)

    # Uniqueness Check
    unique = is_unique(password)

    print("\n--- RESULT ---")
    print("Strength:", strength)

    if not unique:
        print("⚠️ This password was used before (Not Unique)")
    else:
        print("✅ Password is unique")

    if remarks:
        print("\nSuggestions:")
        for r in remarks:
            print("-", r)

    # Suggest Strong Password
    if strength != "Strong":
        print("\n💡 Suggested Strong Password:", generate_password())

    # Store Password
    if unique:
        store_password(password)
        print("\n✅ Password saved securely!")

# Run
main()
```
